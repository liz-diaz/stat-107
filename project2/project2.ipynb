I also wanted to see the differences in size of data and songs if I split it up into low and high ranges of Loudness and Energy to get an idea if there are more or less songs in one range and how that might affect the data science section. The range for loudness is from -46.3 through 0.92 therefore I found the middle point and set the left side as low loudness and the right side as high loudness. As I expected, there are a lot more songs with higher Energy and Loudness:

```python
df_low = df[((df.Energy >= 0) & (df.Energy <= 0.5)) & ((df.Loudness<=-22.69))]
df_low
```

```python
df_high = df[((df. Energy >= 0.5) & (df.Energy <= 1.0)) & (df. Loudness >= -22.69)]
df_high
```

I created samples of random songs from the low and high dataframes since I want an equal number to run an analysis on. I did this by choosing only to look at Loudness and Energy since I want to see if there is any relevant correlation between the two for this project:

```python
dance_energy = ["Loudness", "Energy"]
sample1= df_low.sample(n = 200)
new_low = sample1[dance_energy]
new_low
```

```python
dance_energy = ["Energy", "Loudness"]
sample2= df_high.sample(n = 200)
new_high = sample2[dance_energy]
new_high
```
